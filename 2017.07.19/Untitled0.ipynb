```python
import numpy as np

def findBin(val, binning, begin=0, end=-1): 
    """Find bin index of `val` in the `binning`
    Args:
        val (float): value of initerest
        binning (np.ndarray): sorted array of N bin boundaries for N-1 bins
        begin (int), end (int): helper values of the binary search, never to be used explicitly
    Returns:
        bin number (int) from [-1:N-1] range where N-1 means overflow, -1 is underflow, and None for errors
    """
    # first (user made) call
    if end < begin:
        end = len(binning)
        # Sanity check
    if len(binning) < 1 or end <= begin:
        return None
    # underflow goes to the before-the-first bin
    if val <  binning[begin]:
        return int(begin-1)
    # overlow goes to the next-to-the-last bin
    if val >= binning[end-1]:
        return int(end-1)
    # partition at the middle boundary
    center = int((end-begin-1)/2) + begin
    # got lucky?
    if val >= binning[center] and val < binning[center+1]:
        return center
    # binary search:
    if val < binning[center]:
        return findBin(val, binning, begin, center)
    else:
        return findBin(val, binning, center+1, end)

#binning = np.arange(0,10,0.5)
#print binning
#print [ (v,findBin(v,binning)) for v in [1,2,3,4,5,6,7,8,9] ]

import pandas as pd

def rocMetric(events, binning):
    """
    Args:
        events (np.ndarray): input matrix consisting of three columns
                             'targets' - target values
                             'predictions' - predicted values
                             'weights' - prevalence correction
                                         (e.g. real `target` spectrum should follow x^-4,
                                          while provided values are destributed as x^-1,
                                          hence each should have a weight proportional to x^-3)
        binning  (np.ndarray): sorted array of N bin boundaries for N-1 bins
    Returns:
        dictionary {thresholdBin: (false_positive_rate, true_positive_rate)}
    """
    binnedByTargets = np.array( [ (e['target'],e['weight'],e['prediction'],findBin(e['target'],binning)) for e in events ],
                                dtype  = [('target', np.float),('weight', np.float),('prediction', np.float), ('bin', np.int32)]
                      )

    groupedByTargetBins = pd.DataFrame(binnedByTargets).groupby('bin')

    weightOfTargetsAboveThr = map(lambda thr:
                                  dict([ (bin, sum(np.array(pdf['weight'])*np.array(pdf['prediction']>=thr)))
                                     for (bin,pdf) in groupedByTargetBins
                                  ]),
                                  binning
                              )

    falsePos =  [ sum([weight for (trueBin,weight) in wAbove.items() if trueBin<0 or binning[trueBin] < thr])
                  for (thr, wAbove) in zip(binning, weightOfTargetsAboveThr)
                ]

    truePos  =  [ sum([weight for (trueBin,weight) in wAbove.items() if trueBin>=0 and binning[trueBin] >= thr])
                  for (thr, wAbove) in zip(binning, weightOfTargetsAboveThr)
                ]

    cumsumBelow = [ sum([ sum(pdf['weight']) for (bin,pdf) in groupedByTargetBins if bin<x ]) for x in range(len(binning)) ]
    cumsumAbove = [ sum([ sum(pdf['weight']) for (bin,pdf) in groupedByTargetBins if bin>=x ]) for x in range(len(binning)) ]

    truePos  = [ v/cumsum if v>0 else 0 for (v,cumsum) in zip(truePos,cumsumAbove)  ]
    falsePos = [ v/cumsum if v>0 else 0 for (v,cumsum) in zip(falsePos,cumsumBelow) ]
    
    return dict(zip(binning, zip(falsePos,truePos)));


events = np.array( [(1.,0.1,1.1),(2.,0.01,1.9),(2.3,0.02,2.51),(3.,0.005,3.)],
                    dtype  = [('target', np.float), ('weight', np.float),  ('prediction', np.float)]
         )
x = rocMetric(
        events,
        np.arange(1.5,10,0.5)
    )

print x #df.describe()
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import VectorIndexer
from pyspark.ml.regression import RandomForestRegressor

assemblerCategorical = VectorAssembler(inputCols=
                            ["clct1", "clct2", "clct3", "clct4",
                             "fr1",   "fr2",   "fr3",   "fr4",
                             "rpc1",  "rpc2",  "rpc3",  "rpc4"],
                            outputCol="categoricalFeatures")

categorical = assemblerCategorical.transform(df3)

featureIndexer = VectorIndexer(inputCol="categoricalFeatures", outputCol="indexedCatFeatures", maxCategories=10).fit(categorical)

assemblerContinuous = VectorAssembler(inputCols=
                            ["theta", "st1_ring2",
                             "dPhi12", "dPhi13", "dPhi14", "dPhi23", "dPhi24", "dPhi34",
                             "dTheta12", "dTheta13", "dTheta14", "dTheta23", "dTheta24", "dTheta34",
                             "dPhiS4", "dPhiS4A", "dPhiS3", "dPhiS3A"],
                            outputCol="continuousFeatures")

assembler = VectorAssembler(inputCols=["indexedCatFeatures","continuousFeatures"],outputCol="features")
#assembler = VectorAssembler(inputCols=["continuousFeatures"],outputCol="features")

rf = RandomForestRegressor(labelCol="ptTrueInv", featuresCol="features", numTrees=300)

rfPipeline = Pipeline(stages=[featureIndexer, assemblerContinuous, assembler, rf])
#rfPipeline = Pipeline(stages=[assemblerContinuous, assembler, rf])

(trainingData, testData) = categorical.randomSplit([0.7, 0.3])

rfModel = rfPipeline.fit(trainingData)

rfPredictions = rfModel.transform(testData)

#rfPredictions.show(10)

print(rfModel.stages[1])
```
