## Formulating the problem

A charged particle propagates through the four layers of steel plates returning
magnetic flux of the solenoid and the four interleaving "sensitive" layers of the
detector. The sensitive layers report coordinates and bend angles of the individual
track segments (highlighted with red arrows in the sketch below):

<img src="https://kkotov.github.io/hep/bend4.png" width=600px />

Collected information is used for the track building and particle's momentum assessment.
Multiple scattering effects and non-uniform magnetic field render analytical momentum
assignment impractical. A better solution is building a statistical model regressing
coordinates of the track segments to the particle's momentum. The model can be trained
using either Monte Carlo simulation (as in the current study) or real data where the
target momentum is evaluated with help of another detector subsystem.

This notebook outlines model building and selection using Apache Spark with python (PySpark),
Scikit-learn, Keras (with TensorFlow back-end), and H2O. Since so many people "speak" nothing else
but python when it comes to data analysis, I repeat my originally study performed in R

# Performance metric

The domain-standard trigger turn-on curve, constructed with a certain threshold_pT
in mind, presents the probability that response_pT > threshold_pT as a function
of true_pT (aka. *target*). An example of turn-ons for one of the models look as follows:

<img src="https://kkotov.github.io/emtfPtRegression/2017.02.10/assets/fig/unnamed-chunk-5-1.png" width=600px/>

The main limitation of the turn-on metric is that the model curves
have to be visually compared for a series of thresholds, which a rather
labor-intensive exercise. Moreover, such metric also does not offer a good quantitative
way to account for the spectrum: due to abundance of low true_pT events a sharp
turn-on with a long left tail may do better as well as do worse compared to broad
turn-on with a smaller tail. The Receiver Operating Characteristic (ROC) curve
is free from such limitations, incorporate the spectrum shape (prevalence of
events with low true_pT), and can be quantified
with a simple area-under-the-curve (AUC) parameter.

Before introducing the ROC metric let me start with the spectrum shape and call it
rate. For some given threshold_pT the convolution of turn-on (again, probability
of accepting the event) with the rate in
the range of [0 < true_pT < threshold_pT] gives a number of false-positive events
and in the range of [threshold_pT < true_pT < inf] gives true-positives. These
numbers, normalized by the integrals of the rate in the two regions, give the
true-/false-positive probabilities. Here is a graphical representation of how
true-/false-positive are defined with respect to some threshold:

<img src="https://kkotov.github.io/emtfPtRegression/2017.02.10/convolve.svg" width=500px/>

In this scenario the "natural parameter"
identifying a point on the ROC curve is the threshold_pT. In the code below
I generate the ROC curve running over the grid of threshold_pT defined by the
**binning** array.

```python
import numpy as np

def findBin(val, binning, begin=0, end=-1): 
    """Find bin index of `val` in the `binning`
    Args:
        val (float): value of initerest
        binning (np.ndarray): sorted array of N bin boundaries for N-1 bins
        begin (int), end (int): helper values of the binary search, never to be used explicitly
    Returns:
        bin number (int) from [-1:N-1] range where N-1 means overflow, -1 is underflow, and None for errors
    """
    # first (user made) call
    if end < begin:
        end = len(binning)
        # Sanity check
    if len(binning) < 1 or end <= begin:
        return None
    # underflow goes to the before-the-first bin
    if val <  binning[begin]:
        return int(begin-1)
    # overlow goes to the next-to-the-last bin
    if val >= binning[end-1]:
        return int(end-1)
    # partition at the middle boundary
    center = int((end-begin-1)/2) + begin
    # got lucky?
    if val >= binning[center] and val < binning[center+1]:
        return center
    # binary search:
    if val < binning[center]:
        return findBin(val, binning, begin, center)
    else:
        return findBin(val, binning, center+1, end)

    
import pandas as pd

def rocMetric(events, binning, absoluteRate=False):
    """
    Args:
        events (np.ndarray):   input matrix consisting of three columns
                               'targets' - target values
                               'predictions' - predicted values
                               'weights' - bayes correction: over-/under-sample training
                                           instances according to the given weights
                                           (e.g. real `target` spectrum should follow x^-3,
                                            while provided values are destributed as x^-1,
                                            hence each should have a weight proportional to x^-2)
                                            'weights' is a pure's man approach better to be avoided
        binning  (np.ndarray): sorted array of N bin boundaries for N-1 bins
        absoluteRate:          an optional handle to get absolute (non-normalizad) false positives 
    Returns:
        dictionary {thresholdBin: (false_positive_rate, true_positive_rate)}
    """
    binnedByTargets = np.array( [ (e['target'],
                                   e['weight'],
                                   e['prediction'],
                                   findBin(e['target'],binning)
                                   ) for e in events ],
                                dtype  = [('target', np.float),
                                          ('weight', np.float),
                                          ('prediction', np.float),
                                          ('bin', np.int32)]
                      )

    groupedByTargetBins = pd.DataFrame(binnedByTargets).groupby('bin')

    weightOfTargetsAboveThr = map(lambda thr:
                                  dict([ (bin, sum(np.array(pdf['weight'])*np.array(pdf['prediction']>=thr)))
                                     for (bin,pdf) in groupedByTargetBins
                                  ]),
                                  binning
                              )

    falsePos =  [ sum([weight for (trueBin,weight) in wAbove.items() if trueBin<0 or binning[trueBin] < thr])
                  for (thr, wAbove) in zip(binning, weightOfTargetsAboveThr)
                ]

    truePos  =  [ sum([weight for (trueBin,weight) in wAbove.items() if trueBin>=0 and binning[trueBin] >= thr])
                  for (thr, wAbove) in zip(binning, weightOfTargetsAboveThr)
                ]

    cumsumBelow = [ sum([ sum(pdf['weight']) for (bin,pdf) in groupedByTargetBins if bin<x ]) for x in range(len(binning)) ]
    cumsumAbove = [ sum([ sum(pdf['weight']) for (bin,pdf) in groupedByTargetBins if bin>=x ]) for x in range(len(binning)) ]

    if absoluteRate :
        cumsumBelow = [1]*len(binning)
    
    truePos  = [ v/cumsum if v>0 else 0 for (v,cumsum) in zip(truePos,cumsumAbove)  ]
    falsePos = [ v/cumsum if v>0 else 0 for (v,cumsum) in zip(falsePos,cumsumBelow) ]
    
    return dict(zip(binning, zip(falsePos,truePos)));


# a bit of unit testing is always a good practice
import unittest

class TestSomeBins(unittest.TestCase): 
    def testBins(self):                          
        """findBin should give known result with known input"""
        step = 1/2.
        binning = np.arange(0,10,step)
        self.assertEqual([ (v,findBin(v,binning)) for v in [1,2,3,4,5,6,7,8,9] ], 
                         [ (v,v/step) for v in [1,2,3,4,5,6,7,8,9] ])

    
class TestSomeROCs(unittest.TestCase): 
    def testROCs(self):                          
        """rocMetric should give known result with known input"""
        events = np.array( [(1.,1.1,0.1),    # underflow for a binning starting at threshold=1.5
                            (2.,1.9,0.01),   # count as inefficiency for threshold=2
                            (2.3,2.51,0.02), # place in falsePos for threshold=2.5, truePos otherwise
                            (3.,3.,0.005)],  # place in truePos (on the edge)
                    dtype = [('target', np.float), ('prediction', np.float), ('weight', np.float)]
        )
        binning = np.arange(1.5,3,0.5)  # three bins total
        roc = rocMetric(events, binning)
        result = sorted([ (findBin(thr, binning),(falsePos,truePos)) for (thr,(falsePos,truePos)) in roc.items()],
                        key = lambda x: x[0])
        expect = [(0, (0, (0.01+0.02+0.005)/(0.01+0.02+0.005))),
                  (1, (0, (0.02+0.005)/(0.01+0.02+0.005))),
                  (2, (0.02/(0.1+0.01+0.02), 0.005/0.005))]
        bins     = [ b for (b,(f,t)) in result ]
        falsePos = [ f for (b,(f,t)) in result ]
        truePos  = [ t for (b,(f,t)) in result ]
        self.assertEqual( np.allclose(bins,     [ b for (b,(f,t)) in expect]), True)
        self.assertEqual( np.allclose(falsePos, [ f for (b,(f,t)) in expect]), True)
        self.assertEqual( np.allclose(truePos,  [ t for (b,(f,t)) in expect]), True)

unittest.main(argv=['ignored', '-v'], exit=False)
```

I'll be visually comparing a lot of ROC curves using a universal plotting function below. Feel free to skip this entire block as it is nothing else but a bulky matplotlib visualization code.

```python
def prepareRocPlot(falseTruePos, refThresh = {}, legPos = 'upper left'):
    """
    Function for plotting the ROC curves; all aesthetics is hardcoded
    Args:
        falseTruePos: dictionary associating curve name (appears on the legend) 
                      to the tuple of return value of the rocMetric call and color
        refThresh: dictionary of points {thresholds:color} highlited along the ROC curves
    Returns:
        figure and axises
    """
    plt.close()

    # a bit of hardcoded aesthetics below
    fig, ax0 = plt.subplots(figsize=(10.5, 6), facecolor='white', edgecolor='white')
    ax0.set_ylabel(r"True positive rate", fontsize=18)
    ax0.set_xlabel(r"False positive rate", fontsize=18)
    ax0.set_ylim(0.7, 1.)
    ax0.set_xlim(0.001, 1.)
    ax0.set_xscale('log')
    plt.title("ROC curves comparison", fontsize=20)

    curves = {}
    points = {}
    
    for (name,(roc,color)) in falseTruePos.items():
        falsePos = [ f for (thr,(f,t)) in roc.items()]
        truePos  = [ t for (thr,(f,t)) in roc.items()]
        curves[name] = plt.scatter(falsePos, truePos, s = 30, c = color, edgecolors = color)
        points[name] = []
        binnedROC = { findBin(thr,binning): point for (thr,point) in roc.items() }
        for (thr,col) in refThresh.items():
            xThr = binnedROC[findBin(thr, binning)][0]
            yThr = binnedROC[findBin(thr, binning)][1]
            points[name].append( plt.scatter([xThr], [yThr], s = 30, c = col, edgecolors = col, marker='^') )

    curves.pop(None, None)
    items = curves.values()
    names = curves.keys()

    if points.items() and points.items()[0][1]:
        items.extend( points.items()[0][1] )
        names.extend( [ str(thr)+" GeV threshold position" for thr in refThresh.keys() ] )

    if items:
        plt.legend( items,
                    names,
                    scatterpoints = 1,
                    loc = legPos,
                    ncol = 1,
                    fontsize = 15 )

    return fig, ax0
```

## Univariate linear regression

Let's now perform the simplest non-regularized linear regression that predicts **1/pT** aliased as **ptTrueInv** (~curvature) from difference of $\phi$ coordinates in the first and second layers **dPhi12**. Although, difference in $\phi$ coordinate can be positive and negative, in our pre-processing compression scheme all of the differences are already given by the absolute value while the relative signs are accounted in **dPhiS** surrogate variables

```python
from pyspark.sql.functions import lit, pow, col, log

# Create 1/pT^2 weight column that, combined with the 1/pT original spectrum, can be used
#  to account 1/pT^3 rate suggested in https://kkotov.github.io/emtfPtRegression/2017.03.03/handout.html
# Also accept only the highest quality events that have measuements (coordinates) in each of the four layers (mode==15)
dataframe2 = (dataframe1
                 .withColumn('weight', pow(col('ptTrueInv'),2))
                 .filter(dataframe1.mode == 15)
             )

print "Starting with ", dataframe2.count(), " rows"
```

```python
from pyspark.sql.functions import explode, udf
from pyspark.sql.types import IntegerType, ArrayType
# Following the last response from
# https://stackoverflow.com/questions/34360324/scala-spark-sample-with-replacement-specify-probabilities
# for "samplig" I implement an entire bloody routine, that would otherwise be a true one-liner in R:
#  dataframe2[ sample(seq_len(nrow(dataframe2)), size = 100000, replace = T, prob = dataframe2$weight), ]

createDups = udf(lambda s: range(np.random.poisson(s*10000)), ArrayType(IntegerType()))
dataframe3 = dataframe2.select('*',explode(createDups(dataframe2.weight))).sample(True,0.01,123) #.limit(300000)

print "After reweighting I have ", dataframe3.count(), " rows"
print "Events with pT>150 GeV before reweighting: ", dataframe2[ dataframe2.ptTrue > 150 ].count()
print "Events with pT>150 GeV after reweighting:  ", dataframe3[ dataframe3.ptTrue > 150 ].count()

(trainingData, testData) = dataframe3.randomSplit([0.7, 0.3])
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression

# Spark models consume feature vector that needs to be built from columns of predictors
oneFeature = VectorAssembler(inputCols=["dPhi12"], outputCol="features")

# no regularization for the linear regression yet
linearReg = LinearRegression(maxIter = 10,
                             regParam = 0.0,
                             elasticNetParam = 1.0,
                             standardization = False,
                             labelCol = "ptTrueInv",
#                             weightCol = "weight"
                             )
# I weight-sample the source data in the previous cell rather than directly use the weights column
#  commented out above because among all only the linear model supports such column-based reweighting
#  while I need to put apples to apples when comparing the other "non-reweightable" models to it

# chaining the transformations
lrPipeline = Pipeline(stages=[oneFeature, linearReg])

# Fit the model
linearModel = lrPipeline.fit(trainingData)

# Use it on the test data
linearTest = linearModel.transform(testData)[['ptTrueInv','dPhi12','prediction','weight']].collect() #take(10000)
```

```python
# gosh, in R the construct below would fit into a single subsetting statement, no more comments ...
skim1 = np.array( [ (1./row['ptTrueInv'], 1./row['prediction'], row['dPhi12'], 1)
                    for row in linearTest
                  ], dtype  = [('target', np.float),
                               ('prediction', np.float),
                               ('dPhi12', np.float),
                               ('weight', np.float)]
        )
# Note inverting target and prediction back to the physics scale above

# Let's now look at the fit visually
plt.close()
fig, ax0 = plt.subplots(figsize=(10.5, 6), facecolor='white', edgecolor='white')
ax0.set_ylabel(r"Simulated 1/$p_T$ values (targets)", fontsize=18)
ax0.set_xlabel(r"Measured $\Delta \phi_{12}$ values", fontsize=18)
ax0.set_ylim(0, 1.)
ax0.set_xlim(0, 500)
plt.scatter(skim1['dPhi12'], 1./skim1['target'], s=1, c='#d6ebf2', edgecolors='#8cbfd0', alpha=1.)

x_vals = np.array(ax0.get_xlim())
y_vals = linearModel.stages[1].intercept + linearModel.stages[1].coefficients * x_vals

plt.plot(x_vals, y_vals, linewidth=3.0, c='#0cbfd0', linestyle='--')
plt.title("Fitting a linear model to dPhi12", fontsize=20)
plt.show()
```

The plot above shows the inflexibility of the linear model. The RMSE optimization against **1/pT** target down-weight the contribution from high **pT** or low **dPhi12** rows. As the result, oppositely to the common sense that suggest that straight energetic tracks (**1/pT** ~= 0) should have little bending (**dPhi12** ~= 0), the regression does not cross the coordinate origin and the intercept limits maximum assignable **pT** around 6 GeV. Let's see at some simple metrics below.

```python
# Simplest in-sample performance metrics (explained, for example, in https://kkotov.github.io/emtfPtRegression/2017.01.19/#4):
linearModelRMSE = linearModel.stages[1].summary.rootMeanSquaredError
print("RMSE: %f" % linearModelRMSE)
print("Fraction of variance explained (R-squared): %f" % linearModel.stages[1].summary.r2)
print "Intercept: ", linearModel.stages[1].intercept, ",  slope: ", linearModel.stages[1].coefficients
print("Significance (tValues): %f, %f" % (linearModel.stages[1].summary.tValues[0],linearModel.stages[1].summary.tValues[1]) )
```

Btw., the metrics above wouldn't be accurate if the column-based reweighting is used in this linear model.

Below I'll build the first ROC curve. I'll use it as a reference to compare with another model a bit later.

```python
# consider equally-spaced binning
binning = np.arange(2,200,0.5)

linearROC = rocMetric(skim1, binning)

fig, ax0 = prepareRocPlot({None:(linearROC,'#0cbfd0')}, {5:'red',4:'brown',3:'green'}, legPos='lower right')
ax0.set_xscale('linear')
ax0.set_ylim(0, 1.)
ax0.set_xlim(0, 1.)
plt.title("ROC curve for univariate linear regression", fontsize=20)
plt.show()
```

## Multivariate linear regression

Let's now consider all of the predictors and select the best model. Leave aside best subset selection as well as forward/backward step-/stage-wise regression. Just perform a scan over lasso regularization hyperparameter that automatically suppresses the least important predictors. Don't forget to standardize the predictors for the regularization to treat them on equal footing.

```python
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorIndexer

# put together all categorical predictors
assemblerCategorical = VectorAssembler(inputCols=
                            ['clct1', 'clct2', 'clct3', 'clct4',
                             'fr1',   'fr2',   'fr3',   'fr4',
                             'rpc1',  'rpc2',  'rpc3',  'rpc4',
                             'st1_ring2'],
                            outputCol='categoricalFeatures')

# put together all continuous predictors
assemblerContinuous = VectorAssembler(inputCols=
                            ['theta',
                             'dPhi12',   'dPhi13',   'dPhi14',   'dPhi23',   'dPhi24',   'dPhi34',
                             'dTheta12', 'dTheta13', 'dTheta14', 'dTheta23', 'dTheta24', 'dTheta34',
                             'dPhiS4',   'dPhiS4A',  'dPhiS3',   'dPhiS3A'],
                            outputCol='continuousFeatures')

# index categorical predictors
catFeatureIndexer = VectorIndexer(inputCol='categoricalFeatures', outputCol='indexedCatFeatures', maxCategories=10)

# put all predictors togeteher in one feature vector
assembler = VectorAssembler(inputCols=['indexedCatFeatures','continuousFeatures'], outputCol='features')

# chain transformations in a pipeline
multiLinRegPipeline = Pipeline(stages=[assemblerCategorical, catFeatureIndexer, assemblerContinuous, assembler, linearReg])

# lasso regularization is a particular case of the elastic net with alpha=1
linearReg.setElasticNetParam(1)
linearReg.setStandardization(True)

# define a hyperparameter grid
regParSweep = [1., 1/2., 1/4., 1/16., 1/32., 1/64., 1/128., 1/256., 1/512., 1/1024., 1/2048., 1/4096., 1/8192., 1/16384., 1/32768., 1/65536.]

crossval = CrossValidator(estimator = multiLinRegPipeline,
                          estimatorParamMaps = ParamGridBuilder()
                                                 .addGrid(linearReg.regParam, regParSweep)
                                                 .build(),
                          evaluator = RegressionEvaluator(labelCol='ptTrueInv', metricName='rmse'),
                          numFolds = 10)

cvLinModel = crossval.fit(trainingData)

# Note, the whole block above fits more or less in two lines in R:
# https://kkotov.github.io/emtfPtRegression/2017.01.19/#8
```

```python
# Let's have a look at the scan
print "Average RMSE metric as a function of regularization parameter:"

plt.close()
fig, ax0 = plt.subplots(figsize=(10.5, 6), facecolor='white', edgecolor='white')
ax0.set_ylabel(r"RMSE", fontsize=18)
ax0.set_xlabel(r"Lasso regularization parameter", fontsize=18)
ax0.set_ylim(0.05, 0.2)
ax0.set_xlim(min(regParSweep), 1.)
ax0.set_xscale('log')
ax0.set_yscale('log')

plt.scatter(regParSweep, cvLinModel.avgMetrics, s=100, c='#d6ebf2', edgecolors='#8cbfd0')

# find the best point
indexRMSEmin = min(xrange(len(cvLinModel.avgMetrics)), key = cvLinModel.avgMetrics.__getitem__)

x_vals = [regParSweep[indexRMSEmin], regParSweep[indexRMSEmin]]
y_vals = [0, 10]
plt.plot(x_vals, y_vals, linewidth=3.0, c='#0cbfd0', linestyle='--')
plt.title("Regularization hyperparameter scan", fontsize=20)
plt.show()

print "\nNumbers from the plot above:"
for i,j in zip(xrange(len(regParSweep)), cvLinModel.avgMetrics):
    sys.stdout.write("%f %s %f" % (regParSweep[i], " => ", j))
    if i == indexRMSEmin :
        print " (best)"
    else :
        print ""

## optionally one can look at some technical details
# print cvLinModel.extractParamMap()
# print cvLinModel.getEstimator().getStages()[4].params
# print cvLinModel.bestModel.stages[4].coefficients
```

No surprise the best fit is achieved at low regularization: it is only when model start chasing the noise, the regularization kicks in. Simple linear model does not have much of flexibility to fit the noise with 30 coefficients in millions of events. Therefore model with small regularization is likely to be the optimal one.

On the other hand, the last seven points are more or less on the same level and the best among them seems to be due to a slight fluctuation down. As I want to have more low-significance predictors to be zeroed out, I choose the right-most point on the lower plateau as the regularization parameter of choice.

Let's now play with NN configuration. At this moment I gave up on any interpretability of the NN for good. I also use the best RF ROC metric as a reference. Several models with 0, 1, and 2 hidden layers are shown below. Just a bit of weight-decay regularization is applied to the weights and biases. It makes training stable and improves reproducibility for different random seeds. Also, the target is **1/pT^0.25** to add more weight of the intermediate pT region (~10-20) GeV in RSS and improve ROC performance there.

```python
from keras import regularizers
from keras import initializers
#from keras.layers import Dropout

# I automatize NN model construction as I am going to create lots of models
def buildNN(nInputs, config) :
    model = Sequential()
    # input layer
    model.add(Dense(config[0],
                    input_dim = nInputs,
                    activation = 'relu',
                    kernel_initializer = initializers.random_uniform(seed=1), #random_normal(stddev=0.01, seed=1), #'uniform',
                    kernel_regularizer = regularizers.l2(0.00001),
                    bias_initializer = initializers.random_uniform(seed=1), #'uniform',
                    bias_regularizer = regularizers.l2(0.000001),
                    #activity_regularizer=regularizers.l1(0.01)
                    )
              )
    #model.add(Dropout(0.2))

    # hidden layers
    for nUnits in config[1:-1] :
        model.add(Dense(nUnits,
                        activation = 'relu',
                        kernel_initializer = initializers.random_uniform(seed=1), #'uniform',
                        kernel_regularizer = regularizers.l2(0.00001),
                        bias_initializer = initializers.random_uniform(seed=1), #'uniform',
                        bias_regularizer = regularizers.l2(0.000001),
                        #activity_regularizer=regularizers.l1(0.01)
                        )
                  )
        #model.add(Dropout(0.2))

    # aggregating output layer (no activation needed)
    model.add(Dense(config[-1],
                    kernel_initializer = initializers.random_uniform(seed=1), #'uniform',
                    kernel_regularizer = regularizers.l2(0.00001),
                    bias_initializer = initializers.random_uniform(seed=1), #'uniform',
                    bias_regularizer = regularizers.l2(0.000001),
                    #activity_regularizer=regularizers.l1(0.01)
                    )
              )
    
    model.compile(loss = 'mean_squared_error', optimizer = 'rmsprop')
    return model

# change target to 1/pT^0.25
y_train = trainSet['invSqrt2PtTrue'].values.ravel()
y_test  = testSet ['invSqrt2PtTrue'].values.ravel()

tfModel3 = buildNN(nInputs=len(predictors), config=(10,1))
tfModel3.fit(x_train, y_train, validation_split=0.2, epochs=5)
tfTest3 = 1./tfModel3.predict(x_test)**4
print "RMSE 10+1: ", math.sqrt( tfModel3.evaluate(x_test, y_test, verbose = 0) )

tfModel4 = buildNN(nInputs=len(predictors), config=(10,5,1))
tfModel4.fit(x_train, y_train, validation_split=0.2, epochs=5)
tfTest4 = 1./tfModel4.predict(x_test)**4
print "RMSE 10+5+1: ", math.sqrt( tfModel4.evaluate(x_test, y_test, verbose = 0) )

tfModel5 = buildNN(nInputs=len(predictors), config=(100,50,1))
tfModel5.fit(x_train, y_train, validation_split=0.2, epochs=10)
tfTest5 = 1./tfModel5.predict(x_test)**4
print "RMSE 100+50+1: ", math.sqrt( tfModel5.evaluate(x_test, y_test, verbose = 0) )

tfModel6 = buildNN(nInputs=len(predictors), config=(100,50,50,1))
tfModel6.fit(x_train, y_train, validation_split=0.2, epochs=10)
tfTest6 = 1./tfModel6.predict(x_test)**4
print "RMSE 100+50+50+1: ", math.sqrt( tfModel6.evaluate(x_test, y_test, verbose = 0) )

tfModel7 = buildNN(nInputs=len(predictors), config=(100,50,50,50,1))
tfModel7.fit(x_train, y_train, validation_split=0.2, epochs=15)
tfTest7 = 1./tfModel7.predict(x_test)**4
print "RMSE 100+50+50+50+1: ", math.sqrt( tfModel7.evaluate(x_test, y_test, verbose = 0) )
```
